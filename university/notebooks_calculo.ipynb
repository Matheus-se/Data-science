```python
def i(t):
    return math.exp(-3*t)*t

data = np.array([i(t) for t in np.arange(-0.05, 5, 0.01)])
```

```python
plt.style.use("seaborn")

plt.plot(np.arange(-0.05, 5, 0.01), data, label="i(t)")
plt.plot([-1, 6], [0, 0], linewidth=0.5, label="Ponto zero", color="k")

plt.yticks(np.arange(min(data), max(data)+0.01, 0.03))
plt.ylabel("$\i(t)$")
plt.xlabel('$\/t$')
plt.margins(-0.12, 0.04)
plt.legend()
plt.title("Gráfico da função $\i(t) = e^{-3t}t$")

plt.show()
```
